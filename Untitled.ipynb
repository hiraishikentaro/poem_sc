```python
import pandas as pd
```

```python
columns = ['title', 'text']
poems = pd.DataFrame(columns = ['title', 'text'])
poem_dic = []

for u in urls:
    url = 'http://mico.o.oo7.jp/poem/' + u
    html = request.urlopen(url)
    soup = BeautifulSoup(html)
    try:
        text = soup.find('div', class_ = 'work').get_text()
        title = soup.title.string
        print(title)
        vals = [title, text]
        poem_dic.append(dict(zip(columns, vals)))
    except:
        pass
    
```

```python
poems = poems.append(poem_dic)
```
